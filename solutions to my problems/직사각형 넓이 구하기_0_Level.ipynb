# **직사각형 넓이 구하기**

**2차원 좌표 평면에 변이 축과 평행한 직사각형이 있습니다. 직사각형 네 꼭짓점의 좌표 [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]가 담겨있는 배열 dots가 매개변수로 주어질 때, 직사각형의 넓이를 return 하도록 solution 함수를 완성해보세요.**

```python
def solution(dots):
    x = []
    y = []
    for i in range(0,4):
        #print(dots[i][0])
        #print(dots[i][1])
        x.append(dots[i][0])
        y.append(dots[i][1])
    #print(max(x))    
    #print(min(x))
    answer = (max(x) - min(x)) * (max(y) - min(y))
    return answer
```
