# **문자열 계산하기**

**my_string은 "3 + 5"처럼 문자열로 된 수식입니다. 문자열 my_string이 매개변수로 주어질 때, 수식을 계산한 값을 return 하는 solution 함수를 완성해주세요.**

```python
def solution(my_string):
    temp = my_string.split()
    print(f"{temp}입력값입니다")
    answer = int(temp[0]) #초기값
    for i in range(1,len(temp)):
        #print(f"{i} 인덱스 ")
        print(f"{temp[i]} 벨류")
        if temp[i] == "+":
            answer += int(temp[i+1])
        elif temp[i] == "-":
            answer -= int(temp[i+1])
        else:
            pass
    print(f"{answer} 리턴값입니다")
    return answer
```
