# **삼각형의 완성조건 2**

**선분 세 개로 삼각형을 만들기 위해서는 다음과 같은 조건을 만족해야 합니다.
가장 긴 변의 길이는 다른 두 변의 길이의 합보다 작아야 합니다.
삼각형의 두 변의 길이가 담긴 배열 sides이 매개변수로 주어집니다. 나머지 한 변이 될 수 있는 정수의 개수를 return하도록 solution 함수를 완성해주세요.**

```python
def solution(sides):
    answer=0
    for i in range(1,2001): #원소값이 1000 이므로 최대 2000까지 계산해야함
        if sides[0]+sides[1]>i and sides[0]+i>sides[1] and sides[1]+i>sides[0]:
            #sides의 합이 i 보다 크다 == # i가 가장 긴 변일 경우
            #sides[0]+i>sides[1] and sides[1]+i>sides[0]  #가장 긴 변이 side[0] 이나 side[1] 일경우
            answer+=1
    return answer
```
