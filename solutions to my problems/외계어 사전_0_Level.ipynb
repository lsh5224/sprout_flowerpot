# **외계어 사전**

**PROGRAMMERS-962 행성에 불시착한 우주비행사 머쓱이는 외계행성의 언어를 공부하려고 합니다. 알파벳이 담긴 배열 spell과 외계어 사전 dic이 매개변수로 주어집니다. spell에 담긴 알파벳을 한번씩만 모두 사용한 단어가 dic에 존재한다면 1, 존재하지 않는다면 2를 return하도록 solution 함수를 완성해주세요.**

```python
def solution(spell, dic):
    # 하나라도 존재한다면 1 존재하지않다면 2
    # 존재하는 경우가 하나라도 있으면 답은 1 하나라도 없으면 2
    # 존재하는 경우 하나만 찾으면 된다
    for i in dic: #dic의 요소들을 i로 하나하나 불러온다
        print(f"{set(spell)} set(spell) 값 입니다 {set(i)} i 값입니다 ")
        if not set(spell) - set(i): #if not == 해당 수식이 False 면 return 출력 set(spell) - set(i) 햇을떄 빈 문자열이라면 해당 수식이 진행함 빈문자열이라면 두 요소는 같은 값임 그런데 set(spell) - set(i) 햇을떄 빈 수식이 아니라면 두 요소는 같은 경우가 아님
            return 1
    return 2
```
