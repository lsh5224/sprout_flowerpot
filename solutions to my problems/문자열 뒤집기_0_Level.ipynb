# **문자열 뒤집기**

**문자열 my_string과 정수 s, e가 매개변수로 주어질 때, my_string에서 인덱스 s부터 인덱스 e까지를 뒤집은 문자열을 return 하는 solution 함수를 작성해 주세요.**

```python
def solution(my_string, s, e):
    answer = ''
    temp = my_string[:s]
    temp1 = my_string[s:e+1]
    temp2 = my_string[e+1:]
    print(temp, temp1, temp2)
    result = ''
    for i in range(len(temp1)-1, -1, -1):
        result += temp1[i]
    print(result)
    answer = temp + result + temp2
    return answer
```
