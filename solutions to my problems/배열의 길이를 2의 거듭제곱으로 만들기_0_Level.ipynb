# **배열의 길이를 2의 거듭제곱으로 만들기**

**정수 배열 arr이 매개변수로 주어집니다. arr의 길이가 2의 정수 거듭제곱이 되도록 arr 뒤에 정수 0을 추가하려고 합니다. arr에 최소한의 개수로 0을 추가한 배열을 return 하는 solution 함수를 작성해 주세요.**

```python
def solution(arr):
    answer = arr
    num = len(arr)
    i = 1
    for i in range(0,1001):
        if 2**i >= num:
            temp = 2**i
            break
    print(temp)
    n = temp - num
    
    for i in range(0,n):
        answer += [0]
    return answer
```
