# **OX퀴즈**

**덧셈, 뺄셈 수식들이 'X [연산자] Y = Z' 형태로 들어있는 문자열 배열 quiz가 매개변수로 주어집니다. 수식이 옳다면 "O"를 틀리다면 "X"를 순서대로 담은 배열을 return하도록 solution 함수를 완성해주세요.**

```python
def solution(quiz):
    answer = []
    temp = []
    temp2 = ''
    for i in quiz:
        i += " "
        print(f"{i} 진행중인 수식입니다")
        for j in range(0,len(i)):
            print(f"{i[j]} 진행중인 i의 j 인덱스 값입니다")
            temp2 += i[j]
            if i[j] == " ":
                temp.append(temp2)
                temp2 = ''
                print(f"{temp} temp 에 추가된 값입니다")
        if "- " in temp:
            result = int(temp[0]) - int(temp[2])
            print(f"{result} 진행중인 temp의 결과 값입니다")
            print(f"{result} x-y의 값입니다")
            print(f"{temp[4]} 현재 표기된 수식의 값입니다")
            if result == int(temp[4]):
                answer.append("O")
                temp = []
            elif result != int(temp[4]):
                answer.append("X")
                temp = []
        else:
            result = int(temp[0]) + int(temp[2])
            if result == int(temp[4]):
                answer.append("O")
                temp = []
            elif result != int(temp[4]):
                answer.append("X")
                temp = []
                
    return answer
```
