# **2의 영역**

**정수 배열 arr가 주어집니다. 배열 안의 2가 모두 포함된 가장 작은 연속된 부분 배열을 return 하는 solution 함수를 완성해 주세요.**

```python
def solution(arr):
    answer = []
    if 2 not in arr:
        answer = [-1]
    else:
        for i in range(0, len(arr)):
            #print(arr[i])
            if arr[i] == 2:
                x = i
                break
        for j in range(len(arr)-1,-1,-1):
            #print(f"{j} 인덱스입니다")
            #print(arr[j])
            if arr[j] == 2:
                y = j
                break
        answer = arr[x:y+1]
    return answer
```
