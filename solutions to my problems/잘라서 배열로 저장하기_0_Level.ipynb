# **잘라서 배열로 저장하기**

**문자열 my_str과 n이 매개변수로 주어질 때, my_str을 길이 n씩 잘라서 저장한 배열을 return하도록 solution 함수를 완성해주세요.**

```python
def solution(my_str, n):
    answer = []
    temp = ''
    for i in my_str:
        print(i)
        temp += i
        if len(temp) == n:
            answer.append(temp)
            temp = ''
    if len(temp) == 0:
        answer = answer
    else:
        answer = answer + [temp]
    return answer
```
