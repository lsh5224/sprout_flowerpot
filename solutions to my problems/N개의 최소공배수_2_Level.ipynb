# **N개의 최소공배수**

**두 수의 최소공배수(Least Common Multiple)란 입력된 두 수의 배수 중 공통이 되는 가장 작은 숫자를 의미합니다. 예를 들어 2와 7의 최소공배수는 14가 됩니다. 정의를 확장해서, n개의 수의 최소공배수는 n 개의 수들의 배수 중 공통이 되는 가장 작은 숫자가 됩니다. n개의 숫자를 담은 배열 arr이 입력되었을 때 이 수들의 최소공배수를 반환하는 함수, solution을 완성해 주세요. 
제한 사항
arr은 길이 1이상, 15이하인 배열입니다.arr의 원소는 100 이하인 자연수입니다.
입출력 예
arrresult[2,6,8,14]168[1,2,3]6**

```python
def solution(arr):
    # 두 자연수의 곱 = 최대공약수 X 최소공배수
    # 최소공배수 = 두 수의 곱 / 최대 공약수
    #[ 2 , 3 , 4, 5 ] 의 최소공배수는 ?
    # 2와 3의 최소공배수 = 6
    # 6과 4의 최소공배수 = 12
    # 12와 5의 최소공배수 = 60
    # 2, 3, 4, 5의 최소공배수 = 60
    # (set1.intersection(set2)) 교집합
    answer = 0
    len_arr = len(arr)
    if len_arr == 1:
        answer = arr[0]
    elif len_arr == 2:
        pass
    else:
        for i in arr[1:]:
            print(i)
            x_list = []
            y_list = []
            if answer == 0:
                answer = arr[0] * arr[1]
                for x in range(1,arr[0]+1):
                    if arr[0] % x == 0:
                        x_list.append(x)
                for y in range(1,arr[1]+1):
                    if arr[1] % y == 0:
                        y_list.append(y)
                temp = max(list(set(x_list).intersection(y_list)))
                answer = answer // temp
            else:
                taget = answer * i
                for x in range(1,answer+1):
                    if answer % x == 0:
                        x_list.append(x)
                for y in range(1,i+1):
                    if i % y == 0:
                        y_list.append(y)
                temp = max(list(set(x_list).intersection(y_list)))
                answer = taget // temp
    return answer
```
