# **두 수의 연산값 비교하기**

**연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다.
12 ⊕ 3 = 1233 ⊕ 12 = 312
양의 정수 a와 b가 주어졌을 때, a ⊕ b와 2 * a * b 중 더 큰 값을 return하는 solution 함수를 완성해 주세요.**

```python
def solution(a, b):
    a_result = str(a) + str(b)
    a_result_int = int(a_result)
    b_result = 2*a*b

    if a_result_int >= b_result:
        answer = a_result_int
    else:
        answer = b_result
    return answer
```
