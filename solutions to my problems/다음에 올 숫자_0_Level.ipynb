**등차수열 혹은 등비수열 common이 매개변수로 주어질 때, 마지막 원소 다음으로 올 숫자를 return 하도록 solution 함수를 완성해보세요.**

```python
def solution(common):
    #등비수열인지 등차수열인지부터 구별해야함
    answer = 0
    for idx in range(0,len(common)):
        print(common[idx])
        if common[1] - common[0] == common[-1] - common[-2]: #등차수열인경우
            answer = common[-1] + (common[-1] - common[-2])
        else: #등차수열인경우
            answer = common[-1] * (common[1] // common[0])
            
    return answer
```
