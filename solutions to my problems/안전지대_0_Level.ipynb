# **안전지대**

**다음 그림과 같이 지뢰가 있는 지역과 지뢰에 인접한 위, 아래, 좌, 우 대각선 칸을 모두 위험지역으로 분류합니다.지뢰는 2차원 배열 board에 1로 표시되어 있고 board에는 지뢰가 매설 된 지역 1과, 지뢰가 없는 지역 0만 존재합니다.지뢰가 매설된 지역의 지도 board가 매개변수로 주어질 때, 안전한 지역의 칸 수를 return하도록 solution 함수를 완성해주세요.**

```python
def solution(board):
    n = len(board)
    x_temp = [-1, 1, 0, 0, -1, -1, 1, 1] #좌우상하11시1시7시5시
    y_temp = [0, 0, -1, 1, -1, 1, -1, 1] #좌우상하11시1시7시5시

    #폭탄이 있는 위치 값 추출
    booms = [] # 폭탄의 위치 리스트
    for x in range(n):
        for y in range(n):
            if board[x][y] == 1:
                #print(f'{board[x][y]} 추가될 위치값입니다')
                booms.append((x, y))

    #폭탄의 위치 리스트를 추출한다
    for x, y in booms:
        #print(f'{x} x값, {y} y값')
        for i in range(8):
            nx = x + x_temp[i]
            ny = y + y_temp[i]
            #print(f'{nx} nx값, {ny} ny 값')
            if 0 <= nx < n and 0 <= ny < n: #위의 폭탄터진곳의 범위를 board 에 인덱스를 활용하여 재할당한다.
                board[nx][ny] = 1 # 0 <= nx < n and 0 <= ny < n 는 n*n 배열을 넘어가면 1을 입력안하게함

    #안전지대만 카운팅
    print(board) # 현재 board는 폭탄의 위치와 위험지역을 1로 바꾼 상태 나머지는 0으로 표기된 안전지대
    answer = 0
    for x in range(n): #board의 x 행
        for y in range(n): #board y 열
            if board[x][y] == 0: #board 의 x행의 y열이 0이라면
                answer += 1 #answer에 1을 더한다!

    return answer
```
