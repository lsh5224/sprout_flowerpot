# **정수를 나선형으로 배치하기**

**양의 정수 n이 매개변수로 주어집니다. n × n 배열에 1부터 n2 까지 정수를 인덱스 [0][0]부터 시계방향 나선형으로 배치한 이차원 배열을 return 하는 solution 함수를 작성해 주세요.**

```python
def solution(n):
    X_idx, Y_idx = 0, 0 #좌표값 초기값 정의
    Condition = "right" #진행방향 초기값 정의

   #모든 요소가 0인 n*n 행렬을 만듬
    temp = [] #모든 요소가 0인 n*n 행렬을 만듬
    for i in range(n):
        temp_1 = []
        for j in range(n):
            temp_1.append(0)
        temp.append(temp_1)
        temp_1 = []

    for number in range(1,n*n+1): #1~n*n 만큼의 값 만큼 넣고 진행해야함
        temp[X_idx][Y_idx] = number #해당 좌표에 number값을 넣는다

        if Condition == "right": #오른쪽인경우
            if Y_idx + 1 == n or temp[X_idx][Y_idx + 1] != 0: #Y좌표 +1 이 n이거나 해당 좌표의 요소가 0이 아니라면 방향을 전황해야함(temp의 초기값은 0 이기 때문에 만약 0 이 아니라면 이미 값이 추가되어있는 상태입니다)
                Condition = "down" #진행변경
                X_idx += 1 #다음 차원의 동일한 y값으로 이동
            else:
                Y_idx += 1 #같은 차원의 다른값에 number를 추가

        elif Condition == "down": #아래방향인경우
            if X_idx + 1 == n or temp[X_idx + 1][Y_idx] != 0: #X좌표 +1 이 n이거나 해당 좌표의 요소가 0이 아니라면 방향을 전황해야함(temp의 초기값은 0 이기 때문에 만약 0 이 아니라면 이미 값이 추가되어있는 상태입니다)
                Condition = "left" #진행 변경 #down 은 진행방향으 x좌표에서 +1되는식이기 때문에 x+1값이 0이 아님을 확인해야함
                Y_idx -= 1
            else:
                X_idx += 1 #진향방향이 x좌표의 +1이기때문에 좌표에 1씩 추가해야합니다

        elif Condition == "left": #방향이 왼쪽인경우
            if temp[X_idx][Y_idx - 1] != 0: #왼쪽인경우 y좌표가 0일때 [0-1]은 이미 다운떄 추가되어있는 값이기 때문에 이미 왼쪽방향으로는 모든 값이 추가된 상태여야합니다
                Condition = "up" #진행방향을 변경합니다
                X_idx -= 1 #up이기 떄문ㅔ x갑을 아래에 있는 상태에서 위로가야하기 떄문에 -1 합니다
            else:
                Y_idx -= 1 #왼쪽으로 가야하기떄문에 y값을 -1 씩 변경되게 해야합니다

        elif Condition == "up": #방향이 위쪽인경우
            if temp[X_idx - 1][Y_idx] != 0: #위쪽인 경우 x좌표가 0일떄 0이 아닌 경우는 이미 right 떄 추가된 값이기 떄문에 방향을 변경해야합니다
                Condition = "right"#방향방향 변경
                Y_idx += 1
            else:
                X_idx -= 1

    return temp
```
