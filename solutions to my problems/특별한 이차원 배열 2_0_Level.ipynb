# **특별한 이차원 배열 2**

**n × n 크기의 이차원 배열 arr이 매개변수로 주어질 때, arr이 다음을 만족하면 1을 아니라면 0을 return 하는 solution 함수를 작성해 주세요.**

```python
def solution(arr):
    answer = 0
    temp = []
    for i in range(0,len(arr)):
        for j in range(0,len(arr)):
            if arr[i][j] == arr[j][i]:
                temp += [arr[i][j]]
                answer = 1
            else:
                answer = 0
                break
    return answer
```
