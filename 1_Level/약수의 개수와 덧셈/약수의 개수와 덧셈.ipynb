# **약수의 개수와 덧셈**

**두 정수 left와 right가 매개변수로 주어집니다. left부터 right까지의 모든 수들 중에서, 약수의 개수가 짝수인 수는 더하고, 약수의 개수가 홀수인 수는 뺀 수를 return 하도록 solution 함수를 완성해주세요.**

```python
def solution(left, right):
    answer = 0
    temp = []
    for num in range(left, right+1):
        print(num)
        for i in range(1,num+1):
            if num % i == 0:
                temp.append([i])
                print(temp)
        if len(temp) % 2 == 0:
            answer += num
            temp = []
        else:
            answer -= num
            temp = []
    return answer
```
