# **원하는 문자열 찾기**

**알파벳으로 이루어진 문자열 myString과 pat이 주어집니다. myString의 연속된 부분 문자열 중 pat이 존재하면 1을 그렇지 않으면 0을 return 하는 solution 함수를 완성해 주세요.**

```python
def solution(myString, pat):
    answer = 0
    myString = myString.upper()
    pat = pat.upper()
    if len(myString) >= len(pat):
        if pat in myString:
            answer = 1
    else:
        answer = 0            
    return answer
```
