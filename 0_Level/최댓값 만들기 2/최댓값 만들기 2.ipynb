# **최댓값 만들기 2**

**정수 배열 numbers가 매개변수로 주어집니다. numbers의 원소 중 두 개를 곱해 만들 수 있는 최댓값을 return하도록 solution 함수를 완성해주세요.**

```python
def solution(numbers):
    temp = numbers
    temp.sort()
    num_1 = temp[0] * temp[1]
    num_2 = temp[len(temp)-1] * temp[len(temp)-2]
    if num_1 > num_2:
        answer = num_1
    else:
        answer = num_2
    return answer
```
