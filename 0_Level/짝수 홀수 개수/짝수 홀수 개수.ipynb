# **짝수 홀수 개수**

**정수가 담긴 리스트 num_list가 주어질 때, num_list의 원소 중 짝수와 홀수의 개수를 담은 배열을 return 하도록 solution 함수를 완성해보세요.**

```python
def solution(num_list):
    i_list = []
    j_list = []
    for i in num_list:
        if i % 2 == 0: #짝수
            i_list.append(i)
        else:
            j_list.append(i)
            
    print(i_list)
    print(j_list)
    answer = [len(i_list), len(j_list)]
    return answer
```
