# **더 크게 합치기**

**연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다.
12 ⊕ 3 = 1233 ⊕ 12 = 312
양의 정수 a와 b가 주어졌을 때, a ⊕ b와 b ⊕ a 중 더 큰 값을 return 하는 solution 함수를 완성해 주세요.**

```python
def solution(a, b):
    a = str(a)
    b = str(b)
    int_a = int(a+b)
    int_b = int(b+a)
    if int_a >= int_b:
        answer = int_a
    else:
        answer = int_b 
    return answer
```
