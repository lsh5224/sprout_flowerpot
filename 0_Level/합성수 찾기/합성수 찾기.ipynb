# **합성수 찾기**

**약수의 개수가 세 개 이상인 수를 합성수라고 합니다. 자연수 n이 매개변수로 주어질 때 n이하의 합성수의 개수를 return하도록 solution 함수를 완성해주세요.**

```python
def solution(n):
    #1 n 이하의 수
    #2 n 을 n 이하의 수들로 나누었을때의 0인 것들의 집합 temp
    #3 temp 의 len 이 3이상인것들의 집합 temp2
    #4tmpe2 의 len
    answer = 0
    temp = []
    for i in range(4, n+1): #1
        for j in range(1, i+1):
            if i % j == 0:
                temp.append(i)
        if temp.count(i) >=3:
            answer += 1
    return answer
```
