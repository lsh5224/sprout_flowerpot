# **특별한 이차원 배열 1**

**정수 n이 매개변수로 주어질 때, 다음과 같은 n × n 크기의 이차원 배열 arr를 return 하는 solution 함수를 작성해 주세요.**

```python
def solution(n):
    answer = []
    
    for i in range(0,n):
        temp = []
        for j in range(0,n):
            if i == j:
                temp += [1]
            else:
                temp += [0]
        answer.append(temp)
        temp = []
    return answer
```
