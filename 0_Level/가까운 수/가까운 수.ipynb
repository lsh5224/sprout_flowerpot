# **가까운 수**

**정수 배열 array와 정수 n이 매개변수로 주어질 때, array에 들어있는 정수 중 n과 가장 가까운 수를 return 하도록 solution 함수를 완성해주세요.**

```python
def solution(array, n):
    answer = 0
    array.sort()
    num_list = []
    for i in array:
        num = abs(i - n)
        num_list.append(num)
    print(num_list)
    min_num = min(num_list)
    print(min_num)
    idx = num_list.index(min_num)
    print(idx)
    answer = array[idx]
    return answer
```
