# **숫자 찾기**

**정수 num과 k가 매개변수로 주어질 때, num을 이루는 숫자 중에 k가 있으면 num의  그 숫자가 있는 자리 수를 return하고 없으면 -1을 return 하도록 solution 함수를 완성해보세요.**

```python
def solution(num, k):
    answer = 0
    for idx, number in enumerate(str(num)):
        #print(idx, number)
        if k == int(number):
            temp = idx+1
            print(temp)
            break
        else:
            temp = -1
            print(temp)
            
    answer = temp
    return answer
```
