# **겹치는 선분의 길이**

**선분 3개가 평행하게 놓여 있습니다. 세 선분의 시작과 끝 좌표가 [[start, end], [start, end], [start, end]] 형태로 들어있는 2차원 배열 lines가 매개변수로 주어질 때, 두 개 이상의 선분이 겹치는 부분의 길이를 return 하도록 solution 함수를 완성해보세요.**

```python
def solution(lines):
    answer = 0
    temp = []
    temp1 = []
    temp3 = []
    for i,j in lines:
        #print(i,j)
        for k in range(i,j+1):
            #print(k)
            temp.append(k)
            if len(temp) > j - i:
                temp1.append(temp)
                temp = []
                #print(temp1)
    #print(temp1)            
    for i in range(0,3):
        print(temp1[i][1:])
        print(temp1[i-1][1:])
        for j in temp1[i][1:]:
            print(j)
            if j in temp1[i-1][1:]:
                temp3 += [j]
                print(f'{temp3} 값추가')
    answer = len(list(set(temp3)))
    return answer
```
