# **가위 바위 보**

**가위는 2 바위는 0 보는 5로 표현합니다. 가위 바위 보를 내는 순서대로 나타낸 문자열 rsp가 매개변수로 주어질 때, rsp에 저장된 가위 바위 보를  모두 이기는 경우를 순서대로 나타낸 문자열을 return하도록 solution 함수를 완성해보세요.**

```python
def solution(rsp):
    answer = []
    #가위는 2 바위는 0 보는 5
    for i in rsp:
        if i == "2":
            i = "0"
            answer.append(i)
        elif i == "0":
            i = "5"
            answer.append(i)
        elif i == "5":
            i = "2"
            answer.append(i)
        answer_1 = ''.join(answer)
    return answer_1
```
