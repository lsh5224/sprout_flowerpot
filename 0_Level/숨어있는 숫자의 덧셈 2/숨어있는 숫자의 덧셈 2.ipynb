# **숨어있는 숫자의 덧셈 2**

**문자열 my_string이 매개변수로 주어집니다. my_string은 소문자, 대문자, 자연수로만 구성되어있습니다. my_string안의 자연수들의 합을 return하도록 solution 함수를 완성해주세요.**

```python
def solution(my_string):
    answer = 0
    num_str = ""
    for i in my_string: # i 에 문자열 각각의 값을 할당해줌
        print(f"{i} 현재 i값")
        if i.isdigit():  # 문자가 숫자인 경우라면
            num_str += i  # 숫자 문자열에 추가
            print(f"{num_str} 숫자문자열에 추가된 값")
        else:               # 문자가 숫자가 아닌 경우라면
            if num_str.isdigit():  # 숫자 문자열인 경우에만 더하기
                answer += int(num_str)  # 정수로 변환 후 더하기
                print(f"{num_str} 더할값")
                print(f"{answer} 더해진 값")
            num_str = ""  # 숫자 문자열 초기화
    if num_str.isdigit(): #만약 마지막이 숫자 문자열이라면!
            answer += int(num_str)
    return answer
```
